```python
artists_count = df["artists"].value_counts().head(10)
plt.figure(figsize=(10, 6))
artists_count.plot(kind="bar", color='magenta')
plt.title('Top 10 Most Common Artists Among Top Trending Songs',fontweight='bold')
plt.xlabel('Artist', fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.grid(True)
plt.show()
```

```python
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'])
popularity= df.groupby('snapshot_date')['popularity'].mean()
plt.figure(figsize=(12, 6))
popularity.plot(color='b',label="Populartity")
plt.xlabel('Date Made Available',fontweight="bold")
plt.ylabel('Popularity (Mean)',fontweight="bold")
plt.title('Popularity Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
artist_country = pd.DataFrame({'artist': df["artists"], 'country': df["country"]})
unique_artists_per_country = artist_country.groupby('country')['artist'].nunique().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
unique_artists_per_country.head(10).plot(kind='bar', color='green')
plt.title('Top Countries with Most Unique Artists', fontweight='bold')
plt.xlabel('Country', fontweight='bold')
plt.ylabel('Number of Unique Artists', fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='snapshot_date', y='daily_rank', data=df, color='black')
plt.title('Daily Rank Trend Over Time', fontweight='bold')
plt.xlabel('Snapshot Date', fontweight='bold')
plt.ylabel('Daily Rank', fontweight='bold')
plt.grid(True)
plt.show()
```

```python
#Format will be served as year/month/date
start_date = pd.Timestamp(2024, 1, 1)
end_date = pd.Timestamp(2024, 3, 15)
top_10_songs = (df[df['snapshot_date'].between(start_date, end_date)].groupby('name')['popularity'].mean().nlargest(10).reset_index()['name'])
print("Top 10 Popular Songs from 1st January 2024 to 15th March 2024:")
print(top_10_songs)
```

```python
plt.figure(figsize=(18,8))
sns.boxplot(x="country",y="danceability",data=df,color="magenta")
plt.xlabel("Country", fontweight="bold")
plt.ylabel("Danceability", fontweight="bold")
plt.show()
```

### Danceability across multiple countries.

```python
plt.figure(figsize=(10,6))
sns.histplot(df["energy"],kde=True, color="b")
plt.xlabel("Energy", fontweight="bold")
plt.ylabel("Frequency", fontweight="bold")
plt.grid(True)
plt.show()
```

### Energy levels along with frequency.

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="snapshot_date",y="popularity",color="b",data=df)
plt.xlabel("Date", fontweight="bold")
plt.ylabel("Popularity",fontweight="bold")
plt.title("Trend in Song Popularity Over Time", fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="is_explicit",data=df,color="purple")
plt.xlabel("Explicit", fontweight='bold')
plt.ylabel("No. Songs", fontweight="bold")
plt.title("Explicit vs Non-Explicit Songs", fontweight="bold")
plt.grid(True)
plt.show()
```

### Non-Explicit Songs > Explicit Songs

```python
plt.figure(figsize=(12,9))
sns.scatterplot(x="danceability",y="energy",data=df,color="b")
plt.xlabel("Danceability", fontweight="bold")
plt.ylabel("Energy", fontweight="bold")
plt.title("Relationship between Danceability and Energy", fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,9))
sns.scatterplot(x="danceability",y="popularity",data=df,color="brown")
plt.xlabel("Danceability",fontweight="bold")
plt.ylabel("Popularity",fontweight="bold")
plt.title("Danceability vs Popularity of Songs",fontweight="black")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,9))
average_song_limit=df.groupby("country")["duration_ms"].mean()/60000
average_song_limit.plot(kind="bar",color="orange")
plt.xlabel("Country",fontweight="bold")
plt.ylabel("Time (Minutes)",fontweight="bold")
plt.title("Average Song limit by each Country",fontweight="bold")
plt.grid(True)
plt.show()
```
